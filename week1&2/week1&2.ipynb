```python
# Data wrangling activity 1
# Create a list of 100 random numbers

int_list = [random.randint(1, 100) for x in range(0, 100)]
int_list
```

```python
# Create a new list from this random list, numbers that are divisible by 3.

div_3 = []
for number in int_list:
    if number % 3 == 0:
        div_3.append(number)
    
div_3
    
```

```python
# Calculate the length of these two lists and store the difference in a new variable.

len_int_list = len(int_list)
len_div_3 = len(div_3)
difference = len_int_list - len_div_3
difference 
```

```python
# Using a loop, perform steps 2 and 3 and find the difference variable three times.
difference_list = []
for x in range(0, 3):

    # making two lists
    number_list = [random.randint(1, 100) for x in range(0, 100)]
    div_by_3_list = [x for x in number_list if x % 3 == 0]
    
    # finding the length
    len_number_list = len(number_list)
    len_div_by_3_list = len(div_by_3_list)
    
    # finding the difference
    difference = len_number_list - len_div_by_3_list
    difference_list.append(difference)
    
difference_list
```

```python
# Find the arithmetic ean of these three difference values.

mean_difference = sum(difference_list) / len(difference_list)
mean_difference
```

```python

# remove new lines
chapter_one = chapter_one.replace('\n', '')
chapter_one = chapter_one.replace('\r', '')
chapter_one
```

```python
# remove symbols both alphanumeric and non

for char in chapter_one:
    if char.isalnum() :
        chapter_one = chapter_one.replace('char', ' ')
    elif char == '.':
        chapter_one = chapter_one.replace('.', ' ')
    elif char == ',':
         chapter_one = chapter_one.replace(',', ' ')
    elif char == '"':
         chapter_one = chapter_one.replace('"', ' ')
    elif char == '!':
         chapter_one = chapter_one.replace('!', ' ')
    elif char == '?':
         chapter_one = chapter_one.replace('?', ' ')
    elif char == ';':
         chapter_one = chapter_one.replace(';', ' ')
        
chapter_one
```

```python
#Combine the preceding code into one block, and this time write a separate function where you
# will pass the list generated from dropwhile, and the function will return the whole number 
# contained in the list. As an example, if you pass [1, 2] to the function, it will return 12.
# Make sure that the return type is indeed a number and not a string. Although this task can 
# be achieved using other tricks, we require that you treat the incoming list as a stack in 
# the function and generate the number by reading the individual digits from the stack.

import math
def return_whole_number(number_stack):
    final_number = 0
    for i in range(0, len(number_stack)):
        final_number += (number_stack.pop() * (math.pow(10, i)))
    return final_number
```

```python
# Activity 4
# Import zip_longest from itertools. Create a function to zip header, line and fillvalue=None.

from itertools import zip_longest
```

```python
def csv_reader_parser(header, line):
    zip = zip_longest(header, line, fillvalue = None)
    make_dict = {kv[0]: kv[1] for kv in zip}
    return make_dict
```
