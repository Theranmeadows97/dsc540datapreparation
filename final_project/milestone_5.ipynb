## Load the data in Sqlite

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
mcumovie_data = pd.read_csv("/Users/theranmeadows/Desktop/dsc540datapreparation/final_project/mcumovie_data.csv")
mcumovie_dataR = pd.read_csv("/Users/theranmeadows/Desktop/dsc540datapreparation/final_project/mcumovie_dataR.csv")
mcukills_data = pd.read_csv("/Users/theranmeadows/Desktop/dsc540datapreparation/final_project/mcukills_data.csv")
mcukills_dataT = pd.read_csv("/Users/theranmeadows/Desktop/dsc540datapreparation/final_project/mcukills_dataT.csv")
mcudeath_data = pd.read_csv("/Users/theranmeadows/Desktop/dsc540datapreparation/final_project/mcudeath_data.csv")
mcudeath_dataT = pd.read_csv("/Users/theranmeadows/Desktop/dsc540datapreparation/final_project/mcudeath_dataT.csv")
```

```python
conn = sqlite3.connect("mcu_data.sqlite")
```

```python
mcumovie_data.to_sql("mcumovie_data", conn, if_exists = "replace")
```

```python
mcukills_data.to_sql("mcukills_data", conn, if_exists = "replace")
```

```python
mcudeath_data.to_sql("mcudeath_data", conn, if_exists = "replace")
```

#### MCU Movie data containing film, title, director, producer and tot box office revenue

```python
cursor = conn.cursor()
cursor.execute('''SELECT * from mcumovie_data''')
data = cursor.fetchall()
for row in data:
  print(row)
```

#### MCU Movie data containing the number of times a hero or villian died

```python
mcu_joined_df = mcumovie_data.join(mcukills_data.set_index('film'), on = 'film')
```

#### The combined data from all three data sets joined together by the 'film' column

```python
mcu_joined_df.head()
```

### Visual #1 What is the highest grossing Marvel film of the Infinity Saga?

```python
mcu_joined_df = mcu_joined_df.sort_values(by = 'box office gross', ascending = False)
```

```python
mcu_joined_df.plot.bar(x = 'film', y = 'box office gross')
plt.title("MCU Box Office Gross")
plt.xlabel("Film")
plt.ylabel('Total Box Office Gross ($billion)')
```

### Visual #2 What is the highest on-screen kill count by a hero or villian from the Infinity Saga?

```python
# changing column names to be more accurate and changing column type to 'int'
mcukills_dataT = mcukills_dataT.rename(columns = {'film': 'character'})
mcukills_dataT['total kills'] = mcukills_dataT['total kills'].astype('int')
```

```python
top_10_kills = mcukills_dataT[['character', 'total kills']].sort_values(by = 'total kills', ascending = False).head(10)
top_10_kills
```

```python
top_10_kills.plot.bar(x = "character", y = "total kills")
plt.title('Top 10 highest On-Screen Kills in MCU')
plt.xlabel('Hero/Villian')
plt.ylabel('Kill Count')
```

Because 'Ronan' seems to be a clear outlier in this data, I will remove him and do another visual to show who, after 'Ronan', had
the highest on-screen kill count

### Visualization #4 remove 'Ronan' as an outlier for kills to see a better representation of kills

```python
top_10_kills[1:].plot.bar(x = "character", y = "total kills")
plt.title('Top 9 highest On-Screen Kills in MCU (after "Ronan")')
plt.xlabel('Hero/Villian')
plt.ylabel('Kill Count')
```

This is a much better representation of kills among heroes and villians. Ronan will still be counted as the highest but it
is necessary to se how far back the other characters are from their own kill counts.

### Visual #3 What is the highest on-screen number of deaths of a hero or villian from the Infinity Saga?

Death count is how many times the hero or villian died on screen.

```python
# changing column type to 'int'
mcudeath_dataT['total deaths'] = mcudeath_dataT['total deaths'].astype('int')
```

```python
top_10_deaths = mcudeath_dataT[['characters', 'total deaths']].sort_values(by = 'total deaths', ascending = False).head(10)
top_10_deaths
```

```python
top_10_deaths.plot.bar(x = "characters", y = "total deaths")
plt.title('Top 10 highest On-Screen Deaths in MCU')
plt.xlabel('Hero/Villian')
plt.ylabel('Death Count')
```
