```python
import pandas as pd
```

```python
marvel_main_df = pd.read_csv("/Users/theranmeadows/Desktop/dsc540datapreparation/final_project/marvel-wikia-data.csv")
marvel_main_df.head()
```

## Transformation 1

### Deleting columns that will not be used

```python
marvel_main_df = marvel_main_df[['name', 'ID', 'ALIGN', 'SEX']]
marvel_main_df.head()
```

## Transformation 2

### Changing column names to be more identifiable

```python
marvel_main_df = marvel_main_df.rename(columns = {'name' : 'superhero_name', 'ID' : 'secret_ID?', 'ALIGN' : 'alignment', 'SEX': 'gender'})
marvel_main_df.head()
```

## Transformation 3

### Split superhero_name column to form new column for name

```python
# This line creates a new column that contains evverything in the 'superhero_name' column after the first '('

marvel_main_df['name'] = marvel_main_df['superhero_name'].str.split('(').str[1]
marvel_main_df.head()
```

```python
# This line cleans up the 'superhero_name' column by keeping everything before the first '('

marvel_main_df['superhero_name'] = marvel_main_df['superhero_name'].str.split('(').str[0]
marvel_main_df.head()
```

## Transformation 4

### Remove the ')' and '\\' charaters from the column

```python
marvel_main_df['name'] = marvel_main_df['name'].str.replace(')', '')
marvel_main_df['name'] = marvel_main_df['name'].str.replace('\\', '')
marvel_main_df.head()
```

## Transformation 5

### Make values to be more readable

```python
marvel_main_df['alignment'] = marvel_main_df['alignment'].str.replace('Good Characters', 'good')
marvel_main_df['alignment'] = marvel_main_df['alignment'].str.replace('Neutral Characters', 'neutral')
marvel_main_df['alignment'] = marvel_main_df['alignment'].str.replace('Bad Characters', 'bad')
marvel_main_df.head()
```

```python
marvel_main_df['gender'] = marvel_main_df['gender'].str.replace('Male Characters', 'm')
marvel_main_df['gender'] = marvel_main_df['gender'].str.replace('Female Characters', 'f')
marvel_main_df.head()
```

```python
marvel_main_df['secret_ID?'] = marvel_main_df['secret_ID?'].str.replace('Secret Identity', 'yes')
marvel_main_df['secret_ID?'] = marvel_main_df['secret_ID?'].str.replace('Public Identity', 'no')
marvel_main_df['secret_ID?'] = marvel_main_df['secret_ID?'].str.replace('No Dual Identity', 'no')
marvel_main_df.head()
```
